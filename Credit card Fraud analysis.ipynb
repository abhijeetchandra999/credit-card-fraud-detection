### Machine Learning - Build Fraud Transaction predictive Model

```python
n_data=data.copy()
```

```python
n_data
```

```python
n_data.drop(["first","last","city","cc_num"],axis=1,inplace=True)
```

##### Delete irrelevant columns

### Column Encoding

#### 1. merchant  column

```python
m_d=n_data["merchant"].value_counts().nlargest(10).index
m_d
```

```python
for i in m_d:
    n_data[i]=n_data["merchant"].apply(lambda x: 1 if x == i else 0)
    
    
```

#### 2. category cloumn

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
n_data["encoded_category"]=encoder.fit_transform(n_data["category"])
```

#### 3. Gender column

```python
n_data["gender_encoder"]=encoder.fit_transform(n_data["gender"])
```

#### 4. State column

```python
n_data["stateencoder"]=encoder.fit_transform(n_data["state"])
```

#### 5. job column

```python
e=n_data["job"].value_counts().nlargest(20).index
e
```

```python
for i in e:
    n_data[i]=n_data["job"].apply(lambda x : 1 if x==i else 0)
```

#### 6. transation_hour column

```python
import numpy as np

n_data['hour_sin'] = np.sin(2 * np.pi * n_data['transation_hour'] / 24)
n_data['hour_cos'] = np.cos(2 * np.pi * n_data['transation_hour'] / 24)
```

#### Drop irrelevant columns

```python
n_data.drop(["merchant","category","gender","state","job","transation_hour"],axis=1,inplace=True)
```

### ckeck outliers

```python
sns.boxplot(n_data["amt"])
plt.show()
```

```python
n_data["amt"]=np.log1p(n_data["amt"])
```

```python
n_data
```

#### Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
```

```python
scale_x=scale.fit_transform(x_u)
```

```python
scale_x=pd.DataFrame(scale_x,columns=x.columns)
```

```python
scale_x
```

#### Split the Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(scale_x,y_u,test_size=0.2,random_state=200)
```
